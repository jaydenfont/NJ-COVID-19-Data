```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
# Raw csv pulled from New York Times Github
url = 'https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-states.csv'
df = pd.read_csv(url)
```

```python
df
```

```python
# Isolate data for New Jersey

# loc isolates data based on labels, save as new dataframe
nj = df.loc[df['state'] == "New Jersey"]
nj
```

```python
# Create figure object

figure, ax = plt.subplots()

# Determine days of outbreak

totalDays = nj.shape[0] # get number of rows containing NJ
njLength = np.linspace(1,25,25)
njCases = np.array(nj["cases"])

# Plot NJ cases vs time

ax.plot(njLength, njCases)
ax.set_title("Cases of COVID-19 in New Jersey")
ax.set_xlabel("Day of Outbreak")
ax.set_ylabel("Number of Cases")
xticks = np.arange(1,26,1)
ax.set_xticks(xticks); # semicolon suppresses output of this function
labelText = "Current number of cases = " + str(njCases.max())
box = dict(boxstyle = 'round', facecolor = 'white')
ax.text(9, 8000,labelText,bbox=box)
```
